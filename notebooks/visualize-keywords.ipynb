```python
import sys, os
import pandas as pd
pickle_path = os.path.join("..","data","zfrsoz-test-de.pkl")
df = pd.read_pickle(pickle_path)
df
```

```python
all_top_keywords =[]
num_kwds = 100
for column in df.columns:
    all_top_keywords.append(df[column].sort_values(ascending=False)[0:num_kwds])
df2 = pd.DataFrame(all_top_keywords).transpose()
df2['sum'] = df2.sum(axis=1, skipna=True, numeric_only=True)
df2['median'] = df2.median(axis=1, skipna=True, numeric_only=True)
df2['mean'] = df2.mean(axis=1, skipna=True, numeric_only=True)
#df2.sort_values(by='median', ascending=False)
df2
```
